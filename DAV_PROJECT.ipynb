```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='age', y='friend_count')
plt.title('Friend Count by Age')
plt.xlabel('Age')
plt.ylabel('Friend Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='gender', y='likes_received')
plt.title('Likes Received by Gender')
plt.xlabel('Gender')
plt.ylabel('Likes Received')
plt.ylim(0, 500)  # to reduce skewness by outliers
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='likes', y='likes_received', hue='gender', alpha=0.5)
plt.title('Likes vs Likes Received')
plt.xlabel('Likes')
plt.ylabel('Likes Received')
plt.xlim(0, 1000)
plt.ylim(0, 1000)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample 1000 users randomly
sample_df = df.sample(1000, random_state=2)

plt.figure(figsize=(12, 6))
sns.scatterplot(data=sample_df, x='userid', y='friendships_initiated', alpha=0.6)
plt.title('UserID vs Friendships Initiated')
plt.xlabel('User ID')
plt.ylabel('Friendships Initiated')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample 1000 users randomly for clarity
sample_df = df.sample(1000, random_state=3)

plt.figure(figsize=(12, 6))
sns.scatterplot(data=sample_df, x='userid', y='dob_day', alpha=0.6)
plt.title('UserID vs. Day of Birth')
plt.xlabel('User ID')
plt.ylabel('Day of Birth (1-31)')
plt.yticks(range(1, 32))
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='age', hue='gender', multiple='stack', bins=30)
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample 1000 users randomly for clarity
sample_df = df.sample(1000, random_state=4)

plt.figure(figsize=(12, 6))
sns.scatterplot(data=sample_df, x='userid', y='tenure', alpha=0.6)
plt.title('UserID vs. Tenure')
plt.xlabel('User ID')
plt.ylabel('Tenure (days on platform)')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='gender', y='tenure')
plt.title('Gender vs. Tenure')
plt.xlabel('Gender')
plt.ylabel('Tenure (Days on Platform)')
plt.show()
```
